##### CONSULTA SELECT * FROM 

```python
# Seleccionamos todo el contenido de la tabla (en este caso todo el contenido del DataFrame que definamos en la consulta)

q = """
SELECT * FROM jugadores
"""

psyqldf(q)
```

##### WHERE

```python
# Creo un DataFrame de jugadores que esten valuados por mas de $ 100.000 euros y que su origen sea Brasil

q = """
SELECT first_name, country_of_birth, market_value_in_eur FROM jugadores
WHERE country_of_birth = 'Brazil'
AND market_value_in_eur >= 100000
"""

df_ejercicio = psyqldf(q)
```

##### GROUP BY

```python
# Configuro pandas para mostrar numeros en formato completo
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Creo una tabla que muestre el TOP 10 de valor promedio de mercado de jugadores, agrupado por club actual y ordenado de mayor promedio a menor promedio.

q = """
SELECT 
current_club_name AS club,
AVG(highest_market_value_in_eur) AS promedio_valor
FROM jugadores
GROUP BY 1
ORDER BY promedio_valor DESC
LIMIT 11
"""

top_10_clubes = psyqldf(q)
```

##### HAVING

```python
# Queremos la misma tabla pero solo con clubes con valor promedio menor a 15,000,000 EUR.
# En este caso aplico HAVING y no WHERE ya que el filtro lo aplicamos sobre el valor ya agrupado, cuando aplicamos WHERE en el primer ejercicio lo hacemos directamente sobre la fila cruda que estamos consultando

q = """
SELECT
current_club_name AS club,
AVG(highest_market_value_in_eur) AS promedio_valor
FROM jugadores
GROUP BY 1
HAVING promedio_valor < 15000000
ORDER BY promedio_valor DESC
LIMIT 11
"""

psyqldf(q)
```

##### LEFT JOIN

```python
# Supongamos que partimos del TOP 10 Clubes segun valuacion promedio de los jugadores.
# Ahora lo que quiero es saber como se llama el estadio de cada uno de estos clubes.
# Para eso vamos a hacer un JOIN entre la tabla que creamos top_10_clubes y la tabla de clubes que se importo al inicio del ejercicio.

clubes
```

```python
q = """
SELECT
a.*,
b.stadium_name
FROM top_10_clubes a
LEFT JOIN clubes b
ON a.club = b.name
"""

psyqldf(q)
```
