```python
import pandas as pd
import seaborn as sns
from sqlalchemy import engine 
from pandasql import sqldf

# Función lambda que hace escribir funciones de SQL mucho más fácil
pysqldf = lambda q: sqldf(q, globals())
```

```python
df = df.rename(columns={"Avg. monthly searches": "avg_searches", "Competition (indexed value)": "competition_index"})
```

```python
q = """
SELECT
Keyword,
avg_searches,
Competition,
competition_index
FROM df
WHERE avg_searches >= 500
"""

df_export = pysqldf(q)
```

```python
df_export.to_csv("datos_google_ads_procesados.csv")
```
