### Tratamiento de valores nulos

```python
# Elimino filas que contienen al menos un valor nulo
df_sin_nulos = salaries.dropna()

# Elimino columnas que contienen al menos un valor nulo
df_sin_nulos_columnas = salaries.dropna(axis=1)
```

### Remplazar nulos por la media

```python
df_sin_nulos["salary"].mean() 
```

```python
## Da error porque no es un int (entero) sino que es un strg (cadena de texto)
## Solucion: Parsear la columna salary

## El codigo toma la columna 'salary' que contiene valores de salario en formato cadena
## Elimina cualquier valor no numerico y en una columna nueva 'salary_numeric' incluye los valores numericos en formato float

df_sin_nulos['salary_numeric'] = df_sin_nulos['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)
```

```python
df_sin_nulos['salary_numeric']
```

```python
df_sin_nulos['salary_numeric'].describe()
```

```python
## Armo un BoxPlot para ver graficamente los percentiles

plt.boxplot(df_sin_nulos['salary_numeric'])

# Incluyo etiquetas y titulo
plt.xlabel('Salarios')
plt.ylabel('Valores en USD')
plt.title('Boxplot')

# Lo muestro
plt.show()
```

```python
## Crear un Histograma
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='blue', alpha=0.7)

# Agrego etiquetas y titulo
plt.xlabel('Salarios')
plt.ylabel('Frecuencia')
plt.title('Histograma de salarios')

# Lo muestro
plt.show()
```

```python
# Filtro valores de salarios mayores o iguales a 20000
df_sin_nulos = df_sin_nulos[df_sin_nulos['salary_numeric'] >= 20000]
```

### Reemplazo valores nulos con la media

```python
df_rellenado = df_sin_nulos.fillna(df_sin_nulos['salary_numeric'].mean())
```

```python
df_rellenado
```

```python
df_rellenado.drop('salary', axis=1, inplace=True)
```

```python
df_rellenado
```

### Agrupar y agregar datos

```python
# Supongamos que tienes un DataFrame llamado df con columnas 'title' y 'salary'
# Agrupar por titulo y calcular el promedio y la mediana del salario para cada titulo
df_grouped = df_rellenado.groupby('title')['salary_numeric'].agg(['mean', 'median', 'count']).reset_index()

# Renombrar las columnas para mayor claridad
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']

# Mostrar DataFrame
df_grouped
```
